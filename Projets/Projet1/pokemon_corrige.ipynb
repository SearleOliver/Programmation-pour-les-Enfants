````python
# ══════════════════════════════════════════════════════════════
#  MOTEUR D'AFFICHAGE — Exécute cette cellule une seule fois
#  Tu n'as PAS besoin de modifier ce code !
# ══════════════════════════════════════════════════════════════
from IPython.display import display, HTML, clear_output
import random, json

def afficher_pokemon(pokemon, titre="Mon Pokémon"):
    """Affiche une carte visuelle pour un Pokémon (dictionnaire)."""
    SPRITES = {
        "Électrik": "⚡", "Feu": "🔥", "Eau": "💧", "Plante": "🌿",
        "Psychique": "🔮", "Normal": "⭐", "Glace": "❄️",
        "Combat": "👊", "Poison": "☠️", "Vol": "🦅", "Dragon": "🐉"
    }
    COULEURS = {
        "Électrik": ("#FFF9C4", "#F9A825", "#1a1a00"),
        "Feu":      ("#FFCCBC", "#BF360C", "#fff"),
        "Eau":      ("#BBDEFB", "#0D47A1", "#fff"),
        "Plante":   ("#C8E6C9", "#1B5E20", "#fff"),
        "Psychique":("#F8BBD9", "#880E4F", "#fff"),
        "Normal":   ("#ECEFF1", "#546E7A", "#fff"),
        "Glace":    ("#E0F7FA", "#006064", "#fff"),
        "Combat":   ("#FFCDD2", "#B71C1C", "#fff"),
        "Poison":   ("#E1BEE7", "#4A148C", "#fff"),
        "Vol":      ("#E8EAF6", "#1A237E", "#fff"),
        "Dragon":   ("#EDE7F6", "#4527A0", "#fff"),
    }
    type_poke = pokemon.get("type", "Normal")
    sprite = SPRITES.get(type_poke, "❓")
    bg, accent, txt = COULEURS.get(type_poke, ("#f5f5f5", "#555", "#fff"))
    pv = pokemon.get("pv", 0)
    pv_max = pokemon.get("pv_max", pv)
    pct = max(0, min(100, int(pv / pv_max * 100))) if pv_max > 0 else 0
    bar_color = "#4CAF50" if pct > 50 else "#FF9800" if pct > 25 else "#f44336"
    attaques = pokemon.get("attaques", [])
    atk_html = "".join([
        f'<span style="background:{accent};color:{txt};padding:3px 10px;border-radius:12px;font-size:12px;font-weight:700;margin:3px 3px 0 0;display:inline-block;">'
        f'{a.get("emoji","⚡")} {a["nom"]} ({a["puissance"]})</span>'
        for a in attaques
    ])
    statut = pokemon.get("statut", None)
    statut_html = f'<div style="margin-top:8px;"><span style="background:#7B1FA2;color:#E1BEE7;padding:2px 10px;border-radius:10px;font-size:11px;font-weight:700;">{statut.upper()}</span></div>' if statut else ""
    erreurs = []
    if not isinstance(pokemon.get("nom"), str):   erreurs.append("❌ 'nom' doit être une chaîne de caractères (string)")
    if not isinstance(pokemon.get("pv"), (int,float)):  erreurs.append("❌ 'pv' doit être un nombre entier")
    if not isinstance(pokemon.get("pv_max"), (int,float)): erreurs.append("❌ 'pv_max' doit être un nombre entier")
    if not isinstance(pokemon.get("attaque"), (int,float)): erreurs.append("❌ 'attaque' doit être un nombre entier")
    if not isinstance(pokemon.get("type"), str):  erreurs.append("❌ 'type' doit être une chaîne de caractères")
    if not isinstance(pokemon.get("attaques"), list): erreurs.append("❌ 'attaques' doit être une liste [ ]")
    err_html = ""
    if erreurs:
        err_html = '<div style="background:#FFEBEE;border:2px solid #f44336;border-radius:10px;padding:10px 14px;margin-top:12px;">' + \
                   '<div style="font-weight:900;color:#B71C1C;margin-bottom:6px;">⚠️ Erreurs détectées :</div>' + \
                   "".join(f'<div style="color:#c62828;font-size:13px;">{e}</div>' for e in erreurs) + "</div>"
    html = f"""
    <div style="font-family:'Nunito',sans-serif;background:#1a1a2e;padding:20px;border-radius:16px;max-width:480px;margin:10px 0;">
      <div style="color:#90caf9;font-size:11px;font-weight:700;text-transform:uppercase;letter-spacing:2px;margin-bottom:12px;">📋 {titre}</div>
      <div style="background:{bg};border-radius:14px;padding:18px;position:relative;">
        <div style="display:flex;justify-content:space-between;align-items:flex-start;">
          <div>
            <div style="font-size:22px;font-weight:900;color:#1a1a2e;">{pokemon.get('nom','???')}</div>
            <span style="background:{accent};color:{txt};padding:3px 12px;border-radius:20px;font-size:11px;font-weight:700;text-transform:uppercase;">{type_poke}</span>
          </div>
          <div style="font-size:52px;line-height:1;">{sprite}</div>
        </div>
        <div style="margin:12px 0 6px 0;">
          <div style="display:flex;justify-content:space-between;font-size:12px;font-weight:700;color:#1a1a2e;margin-bottom:4px;">
            <span>❤️ Points de Vie</span><span>{pv}/{pv_max}</span>
          </div>
          <div style="background:rgba(0,0,0,0.15);border-radius:6px;height:14px;overflow:hidden;">
            <div style="width:{pct}%;height:100%;background:{bar_color};border-radius:6px;transition:width 0.5s;"></div>
          </div>
        </div>
        <div style="font-size:13px;font-weight:700;color:#1a1a2e;margin:8px 0 4px 0;">⚔️ Attaque : <strong>{pokemon.get('attaque',0)}</strong></div>
        {statut_html}
        <div style="margin-top:10px;">
          <div style="font-size:11px;font-weight:700;color:rgba(0,0,0,0.5);margin-bottom:4px;">ATTAQUES :</div>
          {atk_html if atk_html else '<span style="color:#999;font-size:12px;">Aucune attaque</span>'}
        </div>
      </div>
      {err_html}
    </div>
    """
    display(HTML(html))


def afficher_combat(joueur, ennemi, log_messages=None):
    """Affiche l'interface de combat entre deux Pokémon."""
    SPRITES = {"Électrik":"⚡","Feu":"🔥","Eau":"💧","Plante":"🌿","Psychique":"🔮","Normal":"⭐","Glace":"❄️","Combat":"👊","Poison":"☠️","Vol":"🦅","Dragon":"🐉"}
    def carte(p, label, color):
        pv = max(0, p.get('pv', 0))
        pv_max = p.get('pv_max', pv)
        pct = max(0, min(100, int(pv/pv_max*100))) if pv_max > 0 else 0
        bar = "#4CAF50" if pct > 50 else "#FF9800" if pct > 25 else "#f44336"
        sprite = SPRITES.get(p.get('type','Normal'), '❓')
        ko_style = "opacity:0.4;" if pv <= 0 else ""
        return f"""
        <div style="flex:1;background:rgba(255,255,255,0.06);border:2px solid {color};border-radius:14px;padding:14px;{ko_style}">
          <div style="font-size:10px;color:{color};font-weight:700;text-transform:uppercase;letter-spacing:1px;">{label}</div>
          <div style="font-size:18px;font-weight:900;color:#fff;margin:4px 0;">{p.get('nom','???')}</div>
          <div style="font-size:40px;text-align:center;margin:6px 0;">{'💀' if pv <= 0 else sprite}</div>
          <div style="font-size:11px;color:#ccc;display:flex;justify-content:space-between;margin-bottom:3px;"><span>PV</span><span>{pv}/{pv_max}</span></div>
          <div style="background:rgba(255,255,255,0.1);border-radius:5px;height:10px;overflow:hidden;">
            <div style="width:{pct}%;height:100%;background:{bar};border-radius:5px;"></div>
          </div>
          <div style="font-size:12px;color:#aaa;margin-top:6px;">⚔️ Attaque : <b style='color:#FFD600;'>{p.get('attaque',0)}</b></div>
        </div>"""
    log_html = ""
    if log_messages:
        CLASSES = {"info":"#90caf9","super":"#FFD600","crit":"#ff8a65","ko":"#ef9a9a","heal":"#a5d6a7","":"#e0e0e0"}
        lignes = "".join(f'<div style="color:{CLASSES.get(cls,"#e0e0e0")};margin-bottom:3px;font-size:13px;">{msg}</div>'
                         for msg, cls in (m if isinstance(m,tuple) else (m,"") for m in log_messages))
        log_html = f'<div style="background:rgba(0,0,0,0.5);border:1px solid rgba(255,255,255,0.1);border-radius:10px;padding:12px 16px;margin-top:12px;min-height:60px;">{lignes}</div>'
    html = f"""
    <div style="font-family:'Nunito',sans-serif;background:#1a1a2e;padding:20px;border-radius:16px;max-width:520px;margin:10px 0;">
      <div style="text-align:center;font-size:11px;color:#90caf9;font-weight:700;letter-spacing:2px;margin-bottom:14px;">⚔️ COMBAT EN COURS</div>
      <div style="display:flex;gap:12px;align-items:stretch;">
        {carte(joueur,'🟢 TOI','rgba(76,175,80,0.6)')}
        <div style="display:flex;align-items:center;font-size:20px;font-weight:900;color:#FFD600;">VS</div>
        {carte(ennemi,'🔴 ENNEMI','rgba(244,67,54,0.6)')}
      </div>
      {log_html}
    </div>
    """
    display(HTML(html))


def afficher_equipe(equipe):
    """Affiche une liste (équipe) de Pokémon sous forme de cartes."""
    SPRITES = {"Électrik":"⚡","Feu":"🔥","Eau":"💧","Plante":"🌿","Psychique":"🔮","Normal":"⭐","Glace":"❄️","Combat":"👊","Poison":"☠️","Vol":"🦅","Dragon":"🐉"}
    cartes = ""
    erreur_liste = ""
    if not isinstance(equipe, list):
        erreur_liste = '<div style="background:#FFEBEE;border:2px solid #f44336;border-radius:10px;padding:10px 14px;color:#B71C1C;font-weight:700;">❌ ton_equipe doit être une liste [ ]</div>'
        display(HTML(f'<div style="font-family:sans-serif;background:#1a1a2e;padding:16px;border-radius:12px;max-width:560px;">{erreur_liste}</div>'))
        return
    for p in equipe:
        t = p.get('type','Normal')
        s = SPRITES.get(t,'❓')
        cartes += f"""
        <div style="flex:1;min-width:120px;background:rgba(255,255,255,0.07);border:1.5px solid rgba(255,255,255,0.15);border-radius:12px;padding:12px;text-align:center;">
          <div style="font-size:34px;">{s}</div>
          <div style="font-size:15px;font-weight:900;color:#fff;margin:4px 0;">{p.get('nom','???')}</div>
          <div style="font-size:11px;color:#FFD600;font-weight:700;">❤️ {p.get('pv',0)} PV</div>
          <div style="font-size:11px;color:#90caf9;">⚔️ {p.get('attaque',0)} ATK</div>
          <div style="font-size:10px;background:rgba(255,255,255,0.1);border-radius:8px;padding:2px 6px;margin-top:6px;color:#ccc;">{t}</div>
        </div>"""
    nb = len(equipe)
    badge = f'<span style="background:#4CAF50;color:#fff;padding:3px 12px;border-radius:20px;font-size:12px;font-weight:700;">{nb} Pokémon</span>'
    html = f"""
    <div style="font-family:'Nunito',sans-serif;background:#1a1a2e;padding:16px;border-radius:14px;max-width:560px;margin:10px 0;">
      <div style="color:#90caf9;font-size:11px;font-weight:700;letter-spacing:2px;margin-bottom:10px;">📋 MON ÉQUIPE {badge}</div>
      <div style="display:flex;gap:10px;flex-wrap:wrap;">{cartes}</div>
    </div>
    """
    display(HTML(html))

print("✅ Moteur d'affichage prêt ! Tu peux maintenant passer aux exercices.")
````

```python

fiche["vitesse"] = 90    # → nouvelle info ajoutée !
```

# 🔴 Exercice 5 — Boucles `for` et `while`

### 📖 Concept : Les boucles répètent du code plusieurs fois.

```python
# Boucle for — parcourir une liste
for pokemon in equipe:
    print(pokemon["nom"])

# Boucle while — répéter TANT QUE une condition est vraie
while joueur["pv"] > 0 and ennemi["pv"] > 0:
    # ... combat ...
```

### ✏️ Ton défi :
1. Change les stats et observe combien de tours dure le combat
2. Ajoute un soin avec `joueur["pv"] += 10` dans la boucle
3. Essaie de faire gagner le joueur en modifiant son attaque !

````python
import random
import time

joueur = {
    "nom":     "Mon Pokémon",
    "type":    "Eau",
    "pv":      35,
    "pv_max":  35,
    "attaque": 14,
}

ennemi = {
    "nom":     "Dracaufeu",
    "type":    "Feu",
    "pv":      38,
    "pv_max":  38,
    "attaque": 16,
}

AVANTAGES = {
    "Feu":      ["Plante"],
    "Eau":      ["Feu"],
    "Électrik": ["Eau"],
    "Plante":   ["Eau"],
}

def update_affichage(joueur, ennemi, messages):
    joueur["pv"] = max(0, joueur["pv"])
    ennemi["pv"] = max(0, ennemi["pv"])
    afficher_combat(joueur, ennemi, messages)
    time.sleep(3)

print(f"⚔️  {joueur['nom']} VS {ennemi['nom']}")
print("═" * 45)

tour = 1

# ══════════════════════════════════════════════════════
#  ✏️ COMPLÈTE la condition du while
#  Indice : le combat continue tant que les deux ont des PV > 0
# ══════════════════════════════════════════════════════
while joueur["pv"] > 0 and ennemi["pv"] > 0:

    print(f"\n  Tour {tour}")

    mult_j = 2.0 if ennemi["type"] in AVANTAGES.get(joueur["type"], []) else 1.0
    mult_e = 2.0 if joueur["type"] in AVANTAGES.get(ennemi["type"], []) else 1.0

    d1 = max(1, int((joueur["attaque"] + random.randint(-3, 3)) * mult_j))
    ennemi["pv"] -= d1

    messages = [
        (f"⚡ {joueur['nom']} attaque !", "info"),
        (f"{'🌟 Super efficace !' if mult_j == 2.0 else ''}", "super"),
        (f"💥 {d1} dégâts infligés", "crit"),
    ]
    update_affichage(joueur, ennemi, messages)

    if ennemi["pv"] <= 0:
        break

    d2 = max(1, int((ennemi["attaque"] + random.randint(-3, 3)) * mult_e))
    joueur["pv"] -= d2

    messages = [
        (f"🔥 {ennemi['nom']} attaque !", "info"),
        (f"{'🌟 Super efficace !' if mult_e == 2.0 else ''}", "super"),
        (f"💥 {d2} dégâts subis", "ko"),
    ]
    update_affichage(joueur, ennemi, messages)

    tour += 1


joueur["pv"] = max(0, joueur["pv"])
ennemi["pv"] = max(0, ennemi["pv"])

# ══════════════════════════════════════════════════════
#  ✏️ COMPLÈTE le résultat final
#  Indice : si le joueur a ses pv > 0 alors victoire !
# ══════════════════════════════════════════════════════

if joueur["pv"] > 0:
    print(f"🏆 {joueur['nom']} GAGNE en {tour} tours !")
    resultat = "victoire"
else:
    print(f"💀 {joueur['nom']} perd après {tour} tours…")
    resultat = "defaite"

msg_fin = [
    (f"⚔️ Combat terminé en {tour} tours !", "info"),
    (f"🏆 Victoire !" if resultat == "victoire" else "💀 Défaite…",
     "super" if resultat == "victoire" else "ko")
]
update_affichage(joueur, ennemi, msg_fin)
````

````python
import random
import time

# ════════════════════════════════════════════════════════════
#  ✏️ ÉTAPE 1 — Crée ton Pokémon champion !
#  Remplace tous les ___ par tes valeurs
# ════════════════════════════════════════════════════════════

mon_champion = {
    "nom":     "Kombatron",   # ← Invente un nom !
    "type":    "Combat",      # ← Quel type bat Psychique ? (indice : regarde AVANTAGES)
    "pv":      45,            # ← Entre 30 et 60
    "pv_max":  45,            # ← Même valeur que pv
    "attaque": 18,            # ← Entre 10 et 25
    "attaques": [
        {"nom": "Poing Éclair",  "puissance": 15, "type": "Combat", "emoji": "👊"},  # indice 0
        {"nom": "Uppercut Final","puissance": 22, "type": "Combat", "emoji": "🥊"},  # indice 1
        # ← Tu peux en ajouter d'autres !                                             # indice 2...
    ]
}

# ════════════════════════════════════════════════════════════
#  ✏️ ÉTAPE 2 — Choisis ton attaque en changeant le chiffre !
#
#  mon_champion["attaques"][0]  →  ta 1ère attaque
#  mon_champion["attaques"][1]  →  ta 2ème attaque
# ════════════════════════════════════════════════════════════

atk_choisie = mon_champion["attaques"][1]  # ← change le chiffre !

# ════════════════════════════════════════════════════════════
#  L'ennemi — Mewtwo (ne pas modifier)
# ════════════════════════════════════════════════════════════
mewtwo = {
    "nom":     "Mewtwo",
    "type":    "Psychique",
    "pv":      50,
    "pv_max":  50,
    "attaque": 20,
    "attaques": [
        {"nom": "Psyko",      "puissance": 22, "type": "Psychique", "emoji": "🔮"},
        {"nom": "Ultralazer", "puissance": 28, "type": "Normal",    "emoji": "💥"},
    ]
}

# ════════════════════════════════════════════════════════════
#  Système de combat — ne pas modifier
# ════════════════════════════════════════════════════════════
AVANTAGES = {
    "Feu":      ["Plante", "Glace"],
    "Eau":      ["Feu", "Roche"],
    "Électrik": ["Eau", "Vol"],
    "Plante":   ["Eau", "Roche"],
    "Psychique":["Combat", "Poison"],
    "Combat":   ["Normal", "Roche", "Glace"],
    "Ténèbres": ["Psychique", "Fantôme"],
    "Insecte":  ["Psychique", "Plante"],
}

def calc_degats(att, type_a, type_d):
    base = att + random.randint(-4, 4)
    if type_d in AVANTAGES.get(type_a, []):
        m = 2.0
    elif type_a in AVANTAGES.get(type_d, []):
        m = 0.5
    else:
        m = 1.0
    if random.random() < 0.1:
        m *= 1.5
    return max(1, int(base * m)), m

print(f"⚔️  {mon_champion['nom']} ({mon_champion['type']}) VS Mewtwo (Psychique)")
print(f"    Attaque choisie : {atk_choisie['emoji']} {atk_choisie['nom']} (puissance {atk_choisie['puissance']})")
print("═" * 50)

tour = 1

while mon_champion["pv"] > 0 and mewtwo["pv"] > 0 and tour <= 20:

    d1, m1 = calc_degats(atk_choisie["puissance"], atk_choisie["type"], mewtwo["type"])
    mewtwo["pv"] = max(0, mewtwo["pv"] - d1)
    eff1 = "🌟 Super efficace !" if m1 >= 2 else "😬 Pas très efficace…" if m1 < 1 else ""

    messages_joueur = [
        (f"🔔 Tour {tour}", "info"),
        (f"{atk_choisie['emoji']} {mon_champion['nom']} utilise {atk_choisie['nom']} !", "info"),
    ]
    if eff1:
        messages_joueur.append((eff1, "super" if m1 >= 2 else "ko"))
    messages_joueur.append((f"💥 {d1} dégâts infligés à Mewtwo", "crit"))

    afficher_combat(mon_champion, mewtwo, messages_joueur)
    time.sleep(2)

    if mewtwo["pv"] <= 0:
        break

    atk_e = random.choice(mewtwo["attaques"])
    d2, m2 = calc_degats(atk_e["puissance"], atk_e["type"], mon_champion["type"])
    mon_champion["pv"] = max(0, mon_champion["pv"] - d2)
    eff2 = "🌟 Super efficace !" if m2 >= 2 else "😬 Pas très efficace…" if m2 < 1 else ""

    messages_mewtwo = [
        (f"🔮 Mewtwo utilise {atk_e['nom']} !", "info"),
    ]
    if eff2:
        messages_mewtwo.append((eff2, "super" if m2 >= 2 else "ko"))
    messages_mewtwo.append((f"💥 {d2} dégâts subis par {mon_champion['nom']}", "ko"))

    afficher_combat(mon_champion, mewtwo, messages_mewtwo)
    time.sleep(2)

    tour += 1

mon_champion["pv"] = max(0, mon_champion["pv"])
mewtwo["pv"]       = max(0, mewtwo["pv"])

if mon_champion["pv"] > 0:
    print(f"\n🏆 VICTOIRE ! {mon_champion['nom']} bat Mewtwo en {tour - 1} tours !")
    msg_fin = [(f"🏆 {mon_champion['nom']} GAGNE !", "super"), (f"Mewtwo vaincu en {tour-1} tours !", "info")]
else:
    print(f"\n💀 DÉFAITE… essaie un autre type ou augmente tes stats !")
    msg_fin = [(f"💀 {mon_champion['nom']} perd…", "ko"), ("💡 Change le type ou augmente les stats !", "info")]

afficher_combat(mon_champion, mewtwo, msg_fin)
````
