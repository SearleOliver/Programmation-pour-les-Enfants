## 📖 C'est quoi une variable ?

Imagine que tu ranges tes affaires dans des boîtes :  
📦 une boîte pour tes crayons, une boîte pour tes cahiers, une boîte pour tes goûters...

En Python, une **variable**, c'est pareil ! C'est une **boîte avec une étiquette** qui contient une valeur.  
Tu lui donnes un nom, et Python retient ce qu'elle contient.

```python
nom_de_la_variable = valeur
```

> 💡 **Attention !** En Python, le signe `=` ne veut **pas dire « égal »**.  
> Il veut dire **« je stocke »** ou **« je mets dans la boîte »** !  
> `age = 12` → "je mets 12 dans la boîte appelée age"

### 👀 Exemple 1 – Mes premières variables

Exécute la cellule ci-dessous ▶️

````python
# On crée 3 variables
prenom = "Alice"
age = 12
ville = "Paris"

# On les affiche
print(prenom)
print(age)
print(ville)
````

Python se souvient de chaque variable jusqu'à la fin du programme !

# 🟡 PARTIE 2 – int, float et opérateurs arithmétiques !
### ⏱️ ~10 minutes

## 📖 Les types numériques

Python fait la différence entre deux sortes de nombres :

| Type | Description | Exemples |
|------|-------------|----------|
| `int` | Nombre **entier** (sans virgule) | `5`, `-3`, `0`, `1000` |
| `float` | Nombre **décimal** (avec virgule) | `3.14`, `-0.5`, `1.0` |

Et les opérateurs disponibles :

| Opérateur | Opération | Exemple | Résultat |
|-----------|-----------|---------|----------|
| `+` | Addition | `5 + 3` | `8` |
| `-` | Soustraction | `10 - 4` | `6` |
| `*` | Multiplication | `3 * 7` | `21` |
| `/` | Division (→ float) | `10 / 4` | `2.5` |
| `//` | Division entière | `10 // 4` | `2` |
| `%` | Reste (modulo) | `10 % 3` | `1` |
| `**` | Puissance | `2 ** 8` | `256` |

### 👀 Exemple 3 – Les opérateurs en action

Exécute la cellule ci-dessous et observe les résultats ▶️

````python
a = 10
b = 3

print("a + b  =", a + b)    # 13
print("a - b  =", a - b)    # 7
print("a * b  =", a * b)    # 30
print("a / b  =", a / b)    # 3.333... (toujours un float !)
print("a // b =", a // b)   # 3 (partie entière)
print("a % b  =", a % b)    # 1 (reste de la division)
print("a ** b =", a ** b)   # 1000 (10 puissance 3)
````

> 💡 **À retenir** : `10 / 3` donne toujours un `float` (`3.333...`).  
> Si tu veux juste la partie entière, utilise `//`.

# 🔴 PARTIE 3 – str et conversions de types !
### ⏱️ ~10 minutes
---

## 📖 Les chaînes de texte (str)

Un `str` (string) est du **texte** entre guillemets. On peut :

| Opération | Syntaxe | Résultat |
|-----------|---------|----------|
| Concaténer (coller) | `"Bon" + "jour"` | `"Bonjour"` |
| Répéter | `"ha" * 3` | `"hahaha"` |
| Longueur | `len("Alice")` | `5` |
| f-string | `f"Bonjour {prenom}"` | `"Bonjour Alice"` |

---

### ⚠️ Les types ne se mélangent pas !

```python
age = 12
print("J'ai " + age + " ans")   # ❌ ERREUR ! on ne peut pas coller str + int
print("J'ai " + str(age) + " ans")   # ✅ on convertit age en texte d'abord
print(f"J'ai {age} ans")             # ✅ les f-strings gèrent ça automatiquement
```

Les 3 fonctions de conversion :
- `int("12")` → convertit un str en entier
- `float("3.14")` → convertit un str en décimal  
- `str(42)` → convertit un nombre en texte

### 👀 Exemple 4 – Jouer avec les str

Exécute la cellule ci-dessous ▶️

````python
prenom = "Alice"
age    = 12

# Méthode 1 : concaténation avec str()
print("Je m'appelle " + prenom + " et j'ai " + str(age) + " ans.")

# Méthode 2 : f-string (recommandée !)
print(f"Je m'appelle {prenom} et j'ai {age} ans.")

# Quelques opérations
print("longueur du prénom :", len(prenom))
print("prénom en majuscules :", prenom.upper())
print("ha" * 5)
````

### 👀 Exemple 5 – La conversion avec input()

> ⚠️ `input()` renvoie **toujours** un `str`, même si l'utilisateur tape un nombre !
