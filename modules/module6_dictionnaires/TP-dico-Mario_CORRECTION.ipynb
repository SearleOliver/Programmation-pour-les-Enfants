    Exercice 8 - Les phrases de Mario

Mario a plusieurs phrase selon si il perd ou si il gagne !
Pour donner de la voix à Mario tu vas créer un dictionnaire phrases
avec les clés "victoire" et "défaite".

Les phrases de victoire de Mario sont :
 - "Here we goo !",

En cas de défaite Mario peut dire :

Petit indice : Rappelle toi de ton cours sur les tableaux

    Exercice 9 - Le combat contre bowser !

Mario arrive devant bowser ! 
Bowser à 4 points de vie il n'a pas suffisament de boule de feu !!
Pour aider Mario tu vas lui ajouter 5 boules de feu dans son inventaire !

Il va lancer ses boules des feu jusqu'à ce qu'il n'en ait plus , pense à les retirer de son inventaire .
Les boule de feu on une chance sur 2 de manquer leur cible , pour savoir si la boule de feu à toucher bowser utilise la fonction **toucher_Bowser(Vie_de_bowser)**


Pour finir tu vas écrire la ligne qui permet d'appeler la phrase de victoire ou de défaite de Mario !

Pour t'aider une partie du code est déjà écrit il te suffit de suivres les instructions avec des # devants !


```python
import random # random signifie aléatoire

def toucher_Bowser(Vie_de_Bowser):
    toucher =random.choice([True, False])
    if toucher:
        Vie_de_Bowser -= 1
        print("Mario touche Bowser !")
    else:
        print("Mario rate sa boule de feu...")
    return Vie_de_Bowser

Vie_de_Bowser = 3

print("Mario arrive devant Bowser ! ")

###### C'est à toi de jouer pour toucher Bowser et le vaincre ! #####




# écris ton code ici
# Ajoute les boules de feu à l'inventaire de Mario
Inventaire["fleurs"] += 5


# écris la boucle qui lance les boule de feu en retirant une boule de feu à chaque passage .
while Inventaire["fleurs"] > 0 and Vie_de_Bowser > 0:
    print("Mario lance une boule de feu ! 🔥")
    Inventaire["fleurs"] -= 1
    Vie_de_Bowser = toucher_Bowser(Vie_de_Bowser)

if Vie_de_Bowser <= 0:
    print("Bowser est vaincu !")
    indice = random.randint(0, 1)
    
    ## écris la ligne qui affiche une phrase de victoire de Mario à l'aide du dictionnaire phrases

    ## petit rappel : "indice" est le nombre aléatoire qui choisit la phrase de victoire à afficher
    print("Mario dit :", phrases["victoire"][indice])


else:
    print("Bowser a encore", Vie_de_Bowser, "points de vie !")
    print("Mario a perdu... ")
    indice = random.randint(0, 1)
    ## écris la ligne qui affiche une phrase de défaite de Mario à l'aide du dictionnaire phrases

    ## petit rappel : "indice" est le nombre aléatoire qui choisit la phrase de défaite à afficher
    print("Mario dit :", phrases["defaite"][indice])
```
