# 🟢 PARTIE 1 – bool et opérateurs de comparaison
### ⏱️ ~10 minutes
---

## 📖 Le type bool

Avant d'utiliser des conditions, il faut connaître le type **bool** (booléen).  
Un booléen ne peut avoir que **deux valeurs** : `True` (vrai) ou `False` (faux).

```python
actif   = True    # comme un interrupteur allumé 💡
termine = False   # comme un interrupteur éteint
```

> ⚠️ **Attention à la majuscule !** En Python, c'est `True` et `False` (pas `true` ni `TRUE`).

---

## 📖 Les opérateurs de comparaison

Ces opérateurs **comparent** deux valeurs et renvoient un `bool` :

| Opérateur | Signification | Exemple | Résultat |
|-----------|---------------|---------|----------|
| `==` | égal à | `5 == 5` | `True` |
| `!=` | différent de | `5 != 3` | `True` |
| `<` | inférieur strict | `3 < 5` | `True` |
| `>` | supérieur strict | `5 > 8` | `False` |
| `<=` | inférieur ou égal | `5 <= 5` | `True` |
| `>=` | supérieur ou égal | `4 >= 6` | `False` |

> 💡 **Ne confonds pas `=` et `==` !**  
> `age = 12` → je stocke 12 dans age (affectation)  
> `age == 12` → je vérifie si age vaut 12 (comparaison)

### 👀 Exemple 1 – Les comparaisons en action

Exécute la cellule ci-dessous ▶️

````python
age = 15

print(age == 15)   # True  — age vaut bien 15
print(age == 18)   # False — age ne vaut pas 18
print(age != 18)   # True  — age est différent de 18
print(age >= 18)   # False — 15 n'est pas >= 18
print(age < 18)    # True  — 15 < 18

# On peut stocker le résultat dans une variable bool !
est_majeur = age >= 18
print("Est majeur ?", est_majeur)
print(type(est_majeur))   # <class 'bool'>
````

Chaque comparaison renvoie `True` ou `False`.  
C'est exactement ce qu'utilise `if` pour décider quoi faire !

# 🟡 PARTIE 2 – if, else et elif
### ⏱️ ~10 minutes
---

## 📖 if, else, elif

Ces trois mots-clés permettent de choisir quel code exécuter selon la situation.

```python
# Structure complète :
if condition_1:
    # exécuté si condition_1 est True
elif condition_2:
    # exécuté si condition_1 est False ET condition_2 est True
elif condition_3:
    # exécuté si les deux précédentes sont False ET condition_3 est True
else:
    # exécuté si TOUTES les conditions au-dessus sont False
```

> 💡 **Règles importantes :**
> - Il faut toujours commencer par `if`
> - `elif` et `else` sont **optionnels**
> - On peut avoir **autant d'elif** qu'on veut
> - Il ne peut y avoir qu'**un seul `else`**, à la fin
> - Python **s'arrête au premier** `if`/`elif` qui est `True`

### 👀 Exemple 2 – if / else

Exécute la cellule ci-dessous ▶️

````python
age = 15

if age >= 18:
    print("✅ Tu es majeur, bienvenue !")
else:
    print("❌ Tu es mineur.")

# Essaie de changer age = 20 et réexécute !
````

### 👀 Exemple 3 – if / elif / else

On peut tester plusieurs cas avec `elif` :

````python
note = 14

if note >= 16:
    print("🏆 Très bien !")
elif note >= 12:
    print("👍 Bien !")
elif note >= 8:
    print("😐 Passable.")
else:
    print("❌ Insuffisant.")

# Essaie avec note = 9, 6, 17 !
````

# 🔴 PARTIE 3 – and, or, not : combiner les conditions !
### ⏱️ ~10 minutes

## 📖 Combiner avec and, or, not

On peut combiner plusieurs conditions pour créer des tests plus complexes.

| Opérateur | Résultat `True` si... | Exemple |
|-----------|----------------------|---------|
| `and` | **les deux** conditions sont `True` | `age >= 18 and a_carte` |
| `or` | **au moins une** condition est `True` | `est_vip or a_invite` |
| `not` | la condition est `False` (inverse) | `not est_ferme` |

```python
# Tables de vérité résumées :
True  and True  → True      True  or True  → True
True  and False → False     True  or False → True
False and True  → False     False or True  → True
False and False → False     False or False → False

not True  → False
not False → True
```

### 👀 Exemple 4 – and et or

Exécute la cellule ci-dessous ▶️

````python
age    = 16
permis = False

# and : les DEUX doivent être True
if age >= 18 and permis:
    print("🚗 Tu peux conduire !")
elif age >= 18 or permis:
    print("⚠️  Une des deux conditions manque.")
else:
    print("❌ Ni l'âge ni le permis.")

# Essaie age=20, permis=True !
````

### 👀 Exemple 5 – not

`not` inverse une condition — très utile pour rendre le code plus lisible !

````python
magasin_ouvert = False

if not magasin_ouvert:
    print("🚪 Le magasin est fermé, reviens plus tard !")
else:
    print("🛒 Tu peux faire tes courses !")

# C'est plus lisible que : if magasin_ouvert == False:
````

### Étape 1 – Une seule question

````python
# ✏️ DÉFI – Étape 1 : une question, une vérification

score = 0

reponse = input("Question 1 : Quel mot-clé crée une variable ? (def / var / = ) : ")

if reponse == "=":
    print("✅ Bonne réponse !")
    score = score + 1
else:
    print("❌ Mauvaise réponse. La bonne réponse était : =")

print("Score actuel :", score, "/ 1")
````

### Étape 2 – Ajoute 2 autres questions !

````python
# ✏️ DÉFI – Étape 2 : le quiz complet à 3 questions

print("=" * 25)
print("   🐍 Quiz Python !")
print("=" * 25)

score = 0

# --- Question 1 ---
print("\nQuestion 1 : Quel est le résultat de 10 % 3 ?")
r1 = input("Ta réponse : ")

if r1 == "1":
    print("✅ Bonne réponse !")
    score += 1
else:
    print("❌ Mauvaise réponse. La bonne réponse était : 1")

# --- Question 2 ---
print("\nQuestion 2 : Que vaut 5 == 5 ?")
r2 = input("Ta réponse : ")

# ✏️ Complète : vérifie si r2 == "True", mets à jour le score



# --- Question 3 ---
print("\nQuestion 3 : Quel mot-clé teste une deuxième condition ?")
r3 = input("Ta réponse (if / elif / else) : ")

# ✏️ Complète : vérifie si r3 == "elif", mets à jour le score



# --- Résultat final ---
print("\n" + "=" * 25)
print("   Résultat final")
print("=" * 25)
print(f"Tu as eu {score} / 3")
````
