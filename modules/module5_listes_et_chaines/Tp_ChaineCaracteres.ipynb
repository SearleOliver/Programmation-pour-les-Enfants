## 📖 Rappel : les indices

Chaque lettre a un **numéro** (index) qui commence à **0** :

```
 D  r  a  c  a  u  f  e  u
[0][1][2][3][4][5][6][7][8]
```

```python
pokemon = "Dracaufeu"

pokemon[5]   # → "u"

```

# 🔥 BADGE FLAMME – La Machine à Évoluer
### *Transforme les noms avec `.upper()`, `.lower()`, `.capitalize()` !*
---

## 📖 Rappel : les méthodes de transformation

```python
pokemon = "salamèche"

pokemon.upper()       # → "SALAMÈCHE"    (TOUT EN MAJUSCULES)
pokemon.lower()       # → "salamèche"    (tout en minuscules)
pokemon.capitalize()  # → "Salamèche"    (1ère lettre en majuscule)
```

> ⚠️ La méthode ne **modifie pas** la variable originale — elle retourne un **nouveau texte** !

````python
# La machine à transformer — exécute et observe !
pokemon = "dracaufeu"

print("Original   :", pokemon)
print("EVOLUTION  :", pokemon.upper())      # crie le nom !
print("minuscule  :", pokemon.lower())      # déjà en minuscule ici
print("Capitalize :", pokemon.capitalize()) # comme dans le Pokédex

# La variable originale n'a pas changé !
print("\nToujours :", pokemon)              # toujours "dracaufeu"
````

### 🌟 Bonus 1 – Le Palindrome Pokémon 🔁

Un **palindrome** se lit pareil dans les deux sens !  
Parmi ces noms de Pokémon (inventés), lesquels sont des palindromes ?

> 💡 `mot[::-1]` inverse un texte en Python — tu verras les slices plus tard !

````python
noms = ["RADAR", "PIKACHU", "KAYAK", "RONFLEX", "LAVAL", "ABBA"]

print("🔍 Détecteur de Palindromes Pokémon")
for nom in noms:
    inverse = nom[::-1]    # inverse le texte (astuce avancée !)
    if nom == inverse:
        print("  🔁", nom, "→ PALINDROME !")
    else:
        print("  ❌", nom, "→ pas un palindrome  (inversé :", inverse + ")")
````
