## 📖 C'est quoi une fonction ?

Imagine que chaque matin tu fais les mêmes gestes :  
👉 tu te lèves → tu t'habilles → tu prends ton sac  

Si tu devais **réécrire** ces étapes chaque jour dans ton agenda, ce serait long ! 😩  
Alors tu écris juste : **"routine du matin"** et tu sais ce que ça veut dire.

En Python, une **fonction**, c'est pareil !  
Tu crées une action avec **`def`**, tu lui donnes un nom, et tu l'appelles quand tu veux.

```python
def nom_de_la_fonction():
    # ce que fait la fonction
    print("Coucou !")

# On l'appelle comme ça :
nom_de_la_fonction()
```

> 💡 **Attention à l'indentation !** Le code à l'intérieur de la fonction doit être **décalé de 4 espaces** (ou une tabulation).

### 👀 Exemple 1 – Regarde comment ça marche

Exécute la cellule ci-dessous ▶️

````python
# On définit la fonction
def dire_bonjour():
    print("👋 Bonjour tout le monde !")

# On l'appelle 3 fois !
dire_bonjour()
dire_bonjour()
dire_bonjour()
````

Tu vois ? On a écrit `dire_bonjour()` **3 fois** et la fonction s'est exécutée 3 fois !  
Imagine si la fonction avait 100 lignes de code... ça économise beaucoup d'écriture 😎

# 🟡 PARTIE 2 – Les paramètres : personnaliser sa fonction !
### ⏱️ ~10 minutes

## 📖 C'est quoi un paramètre ?

Imagine une machine à smoothie 🥤 :  
- Si tu mets des **fraises** → elle fait un smoothie à la fraise  
- Si tu mets des **bananes** → elle fait un smoothie à la banane  

La machine est toujours **la même** (la fonction), mais ce qu'on met dedans (le **paramètre**) change le résultat !

```python
def faire_smoothie(fruit):          # 'fruit' est le paramètre
    print("🥤 Smoothie à la " + fruit)

faire_smoothie("fraise")            # on passe "fraise" comme argument
faire_smoothie("banane")            # on passe "banane" comme argument
```

### 👀 Exemple 2 – La fonction salutation personnalisée

Exécute la cellule ci-dessous ▶️

````python
def saluer(prenom):
    print("🌟 Bonjour " + prenom + " ! Bienvenue en Python !")

saluer("Alice")
saluer("Bob")
saluer("Rania")
````

La même fonction `saluer`, mais elle dit bonjour à **des personnes différentes** !  
C'est la magie des paramètres. 🪄

### 👀 Exemple 3 – Plusieurs paramètres à la fois !

On peut avoir **plusieurs paramètres**, séparés par des virgules :

````python
def carte_identite(prenom, age, ville):
    print("📋 Fiche élève :")
    print("   Prénom : " + prenom)
    print("   Âge    : " + str(age) + " ans")
    print("   Ville  : " + ville)
    print("---")

carte_identite("Alice", 12, "Paris")
carte_identite("Bob", 11, "Lyon")
````

> 💡 **Remarque** : `str(age)` transforme le nombre `12` en texte `"12"` pour pouvoir le coller avec du texte. Sans ça, Python ferait une erreur !

# 🔴 PARTIE 3 – Le `return` : la fonction qui répond !
### ⏱️ ~10 minutes
---

## 📖 C'est quoi le `return` ?

Jusqu'ici, nos fonctions **affichaient** des résultats avec `print()`.  
Mais parfois, on veut que la fonction **calcule quelque chose et nous le renvoie** pour pouvoir le réutiliser.

| | `print()` | `return` |
|---|---|---|
| **Ce que ça fait** | Affiche à l'écran | Renvoie une valeur |
| **Peut-on la réutiliser ?** | ❌ Non | ✅ Oui |
| **Exemple** | `print(5 + 3)` → affiche `8` | `return 5 + 3` → renvoie `8` |

```python
def doubler(nombre):
    return nombre * 2          # la fonction renvoie le résultat

resultat = doubler(5)          # on stocke le résultat dans une variable
print(resultat)                # affiche : 10
```

### 👀 Exemple 4 – Calcule et renvoie !

Exécute la cellule ci-dessous ▶️

````python
def additionner(a, b):
    return a + b

resultat = additionner(3, 5)
print("3 + 5 =", resultat)

# On peut même utiliser le résultat dans un autre calcul !
double = additionner(resultat, resultat)
print("Le double de", resultat, "c'est", double)
````

### 👀 Exemple 5 – La différence entre `print` et `return`

Regarde bien la différence !

````python
# Avec print() → affiche mais ne renvoie rien d'utile
def multiplier_print(a, b):
    print(a * b)

# Avec return → renvoie le résultat qu'on peut utiliser
def multiplier_return(a, b):
    return a * b

# Test :
x = multiplier_print(4, 3)    # affiche 12 mais x vaut None !
print("x vaut :", x)          # x = None  😬

print("---")

y = multiplier_return(4, 3)   # y vaut 12 ✅
print("y vaut :", y)          # y = 12  🎉
````
