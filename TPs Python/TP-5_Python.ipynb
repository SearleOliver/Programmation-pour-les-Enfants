## üßÆ Partie 1 ‚Äì Les puissances et les racines carr√©es en Python

En Python, pour faire une puissance (comme 5¬≤), on utilise ** (deux √©toiles) :

```python
print(3**2)  # donne 9

print(2**4)  # donne 16
```

Et pour faire une racine carr√©e, on utilise sqrt() qui vient du module math :

```python
from math import sqrt

print(sqrt(25))  # donne 5
print(sqrt(9))   # donne 3
```

üì¶ Cette ligne : from math import sqrt

sert √† importer la fonction sqrt (pour faire une racine carr√©e) depuis la biblioth√®que math de Python.

üß† Une biblioth√®que, c‚Äôest comme une bo√Æte √† outils d√©j√† pr√™te, qui contient des fonctions utiles que Python ne conna√Æt pas tout seul.

### ‚ö†Ô∏è Attention aux erreurs !

üß† En Python (comme en maths), on ne peut pas calculer la racine carr√©e d‚Äôun nombre n√©gatif.

Sinon, Python affiche une erreur de domaine math√©matique (Math domain error).

```python
from math import sqrt

print(sqrt(-5))
```

üí• R√©sultat :

#### ValueError: math domain error

üö´ Cela veut dire : ‚ÄúTu me demandes un calcul qui n‚Äôexiste pas en maths classiques !‚Äù

‚úÖ Astuce : Avant de faire une racine carr√©e, tu peux v√©rifier que le nombre est positif ou nul :

```python
from math import sqrt

x = int(input("Veuilez rentrer un nombre afin de calculer sa racine carr√©e : "))

if x >= 0:
    print(sqrt(x))
    
else:
    print("Impossible de calculer la racine d‚Äôun nombre n√©gatif.")
```

## üéÆ Partie 2 ‚Äì Jeu de calcul mental

Voici un petit jeu pour t'entra√Æner avec des additions, soustractions et multiplications.


Tu dois ajouter la division, et faire attention √† ne pas diviser par 0 !

```python
import random

a = random.randint(1, 10)
b = random.randint(1, 10)

# on choisit une op√©ration au hasard
operation = random.choice(['+', '-', '*']) # Rajouter ici le symbole division qui est : /

# on calcule le bon r√©sultat
if operation == '+':
    result = a + b
elif operation == '-':
    result = a - b
# elif rajouter la condition pour la division  
else:
    result = a * b


print(f"Combien font {a} {operation} {b} ?")
reponse = int(input("> "))


if reponse == result:
    print("Bonne r√©ponse ! üéâ")
else:
    print(f"Mauvaise r√©ponse ! La bonne r√©ponse √©tait {result}. üòÖ")
```
