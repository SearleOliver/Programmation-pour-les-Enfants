## üì¶ Partie 1 ‚Äì C‚Äôest quoi turtle ?

`Turtle` est une biblioth√®que Python qui te permet de cr√©er des dessins en donnant des instructions √† une petite tortue magique  üê¢.


Elle se d√©place sur un fond qui ressemble √† un plan de maths (le plan cart√©sien), et tu peux lui dire d‚Äôavancer, de tourner, ou de tracer des formes. üé®

On d√©part, la tortue est plac√©e au point de coordonn√©es (0,0), au centre de la fen√™tre graphique, orient√©e vers la droite.

La tortue est capable de r√©aliser un certain nombre de commandes 

On d√©part, la *tortue* est plac√©e au point de coordonn√©es `(0,0)`, au centre de 
la fen√™tre graphique, orient√©e vers la droite.

La *tortue* est capable de r√©aliser un certain nombre de commandes :

**Commande**         | **Description**
---------------------|----------------------------------------------------
`speed(vitesse)`     | Change la vitesse de dessin (de 0 √† 10)
`up()`               | L√®ve le stylo (ne dessine plus)
`down()`             | Abaisse le stylo (dessine √† nouveau)
`goto(x,y)`          | Se d√©place aux coordonn√©es `(x,y)`
`forward(distance)`  | Avance de `distance` selon sa direction actuelle
`backward(distance)` | Recule de `distance` selon sa direction actuelle
`right(angle)`       | Tourne √† droite de `angle` degr√©s
`left(angle)`        | Tourne √† gauche de `angle` degr√©s
`begin_fill()`       | Commence le remplissage d‚Äôune forme
`end_fill()`         | Termine et remplit la forme
`color(couleur)`     | Change la couleur (`'red'`, `'blue'`, `'black'`, ...)
`done()`             | Termine le programme et maintient la fen√™tre ouverte

### üü¶ 1) Le carr√© magique

#### üá≥üá± Pays-Bas (exemple donn√©)

```python
from turtle import *
speed(0)

largeur_drapeau = 300
hauteur_drapeau = 210

up()
goto(-largeur_drapeau//2,-hauteur_drapeau//2)
down()

left(90)

begin_fill()
color('blue')
forward(hauteur_drapeau//3)
right(90)
forward(largeur_drapeau)
right(90)
forward(hauteur_drapeau//3)
right(90)
forward(largeur_drapeau)
right(90)
end_fill()

forward(hauteur_drapeau//3)

begin_fill()
color('white')
forward(hauteur_drapeau//3)
right(90)
forward(largeur_drapeau)
right(90)
forward(hauteur_drapeau//3)
right(90)
forward(largeur_drapeau)
right(90)
end_fill()

forward(hauteur_drapeau//3)

begin_fill()
color('red')
forward(hauteur_drapeau//3)
right(90)
forward(largeur_drapeau)
right(90)
forward(hauteur_drapeau//3)
right(90)
forward(largeur_drapeau)
right(90)
end_fill()
# indispensable pour obtenir un affichage et sortir proprement !
done()
try:
    bye()
except:
    print("bye")
```

#### üá´üá∑ France (exemple donn√©)

```python
from turtle import *

speed(0)
width(1)
penup()

# Param√®tres du drapeau
largeur = 300
hauteur = 200
x_depart = -largeur // 2
y_depart = hauteur // 2

# Fonction pour dessiner une bande verticale
def bande(couleur):
    begin_fill()
    color(couleur)
    for _ in range(2):
        forward(largeur // 3)
        right(90)
        forward(hauteur)
        right(90)
    end_fill()

# Dessin des bandes
goto(x_depart, y_depart)
setheading(0)  # face √† droite
pendown()
bande("blue")

penup()
forward(largeur // 3)
pendown()
bande("white")

penup()
forward(largeur // 3)
pendown()
bande("red")

hideturtle()
done()
```
